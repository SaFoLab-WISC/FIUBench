```python
from apis.gemini import GeminiEvaluator
from collections import defaultdict
import json
import os

def load_ISEKAI_image(root):
    def zero(): return 0
    example_num = defaultdict(zero)
    examples = []

    labels = os.listdir(root)
    labels = [item for item in labels if "." not in item]
    img_num = 0
    for label in labels:
        for category in os.listdir(os.path.join(root, label)):
            if "pos" in category:
                path = os.path.join(root, label, category)
                for img_name in os.listdir(path):
                    if ".DS_Store" in img_name: continue
                    img_path = os.path.join(path, img_name)
                    example_num[label] += 1
                    examples.append({
                        "image_list": [img_path],
                        "label": label,
                    })
                    if example_num[label] >= 5:
                        break
                    
    with open("/mnt/d/datasets/vlm_unlearned/dataset/vlm_unlearned_examples.json", "w") as f:
        for line in examples:
            f.write(f"{json.dumps(line)}\n")
            
    return examples

root = "/mnt/d/datasets/vlm_unlearned/ISEKAI/ISEKAI-pair"
examples = load_ISEKAI_image(root)
print(len(examples))
```
